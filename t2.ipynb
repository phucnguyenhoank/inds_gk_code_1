```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Generate random data
np.random.seed(10)
data = np.array([3, 4, 5, 5, 7, 8, 9, 10, 11, 11, 12])

# Convert to Pandas Series
data_series = pd.Series(data)

# Now describe() works
print(data_series.describe())

# Create box plot
plt.figure(figsize=(6, 4))
sns.boxplot(data=data)

# Set y-axis ticks at whole numbers
plt.yticks(np.arange(0, max(data) + 1, 1))

# Add labels
plt.title("Box Plot Example")
plt.ylabel("Value")

# Show plot
plt.show()
```
